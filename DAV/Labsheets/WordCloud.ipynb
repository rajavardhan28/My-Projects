```python
text=" ".join(item for item in df['cast'])
print(text)
```

```python
stopwords=set(STOPWORDS)
```

```python
wordcloud=WordCloud(background_color="yellow").generate(text)
plt.imshow(wordcloud,interpolation="bilinear")
plt.axis("off")
plt.margins(x=0,y=0)
plt.show()
```

```python
df=pd.read_csv('/content/7817_1.csv',usecols=['reviews.username'])
df.head()
```

```python
ndf=df.dropna()
```

```python
text=" ".join(item for item in ndf['reviews.username'])
print(text)
```

```python
wordcloud=WordCloud(background_color="aqua").generate(text)
plt.imshow(wordcloud,interpolation="bilinear")
plt.axis("off")
plt.margins(x=0,y=0)
plt.show()
```

```python
wordcloud=WordCloud(background_color="beige",
                    max_words=100,
                    max_font_size=300,
                    width=800,
                    height=500,
                    colormap="plasma"
                    ).generate(text)
plt.figure(figsize=(20,20))
plt.imshow(wordcloud,interpolation="hamming")
plt.axis("off")
plt.margins(x=0,y=0)
plt.savefig("clouds.jpg",format="jpg")
plt.show()
```
