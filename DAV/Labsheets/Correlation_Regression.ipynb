```python
from scipy import stats
```

```python
x=[5,7,8,7,2,17,2,9,4,11,12,9,6]
y=[99,86,87,88,111,86,103,87,94,78,77,85,86]
```

```python
slope,intercept,r,p,std_err=stats.linregress(x,y)
```

```python
def myfunc(x):
  return slope * x + intercept
```

```python
mymodel=list(map(myfunc,x))
```

```python
plt.scatter(x,y)
```

```python
plt.plot(x,mymodel)
```

```python
import numpy as np
import matplotlib.pyplot as mtplt

def estimate_coeff(p,q):
  n1=np.size(p)
  m_p=np.mean(p)
  m_q=np.mean(q)
  SS_pq=np.sum(q*p)-n1*m_q*m_p
  SS_pp=np.sum(p*p)-n1*m_q*m_p
  b_1=SS_pq/SS_pp
  b_0=m_q-b_1*m_p
  return(b_0,b_1)
def  plot_regression_line(p,q,b):
  mtplt.scatter(p,q,color="m",marker="o",s=30)
  q_pred=b[0]+b[1]*p
  mtplt.plot(p,q_pred,color="g")
  mtplt.xlabel('p')
  mtplt.ylabel('q')
  mtplt.show()
def main():
  p=np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
  q=np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
  b=estimate_coeff(p,q)
  print("Estimated coefficients are:\nb_0 = () \ \nb_1=()".format(b[0],b[1]))
  plot_regression_line(p,q,b)
if __name__=="__main__":
  main()
```
